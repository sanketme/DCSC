```python
data = vmodel.run(return_columns=['CCCM', 'CCCW', 'CCCR','production rate', 'OrderRateW','OrderRateR', 'wholesaler amount to order WOP', 'retailer amount to order ROP', 'Offline Customer Order' ])
```

```python
var_data = data.var()
```

```python
CFB_M = var_data['CCCM'] / var_data['wholesaler amount to order WOP']
CFB_W = var_data['CCCW'] / var_data['retailer amount to order ROP']
CFB_R = var_data['CCCR'] / var_data['Offline Customer Order']
BWE_M = var_data['production rate'] / var_data['wholesaler amount to order WOP']
BWE_W = var_data['OrderRateW'] / var_data['retailer amount to order ROP']
BWE_R = var_data['OrderRateR'] / var_data['Offline Customer Order']
print('CFB_M= ', CFB_M)
print('CFB_W= ', CFB_W)
print('CFB_R= ', CFB_R)
print('BWE_M= ', BWE_M)
print('BWE_W= ', BWE_W)
print('BWE_R= ', BWE_R)
```
